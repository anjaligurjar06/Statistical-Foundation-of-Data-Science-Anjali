Importing necessary libraries and Dataset we used in Practical 2

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/teachers_rating_an.csv")
df.head(10)
```

Q1) Identify all duplicate cases using prof. Using all observations, find the average and standard deviation for age. Repeat the analysis by first filtering the data set to include one observation for each instructor with a total number of observations restricted to 94.

```python
duplicates = df[df.duplicated(subset='prof', keep=False)]
print(duplicates.head(), "\n")

mean_age_all = df['age'].mean()
std_age_all = df['age'].std()
print(f"Mean Age (All Observations): {mean_age_all:.2f}")
print(f"Std Dev Age (All Observations): {std_age_all:.2f}\n")

df_unique = df.drop_duplicates(subset='prof', keep='first').head(94)
mean_age_unique = df_unique['age'].mean()
std_age_unique = df_unique['age'].std()
print(f"Mean Age (Unique Professors): {mean_age_unique:.2f}")
print(f"Std Dev Age (Unique Professors): {std_age_unique:.2f}\n")
```

Q2) Using a bar chart, demonstrate if instructors teaching lower-division courses receive higher average teaching evaluations.

```python
import matplotlib.pyplot as plt
import seaborn as sns


eval_by_division = df.groupby('division')['eval'].mean().reset_index()

# Bar chart
sns.barplot(x='division', y='eval', data=eval_by_division, color='#ffb6c1')  # light pink
plt.title("Average Teaching Evaluation by Course Division", fontsize=14, fontweight='bold', color='#cc0066')
plt.xlabel("Course Division", fontsize=12, color='#660033')
plt.ylabel("Average Evaluation", fontsize=12, color='#660033')
plt.show()
```

Q3) Plot the relationship between age and teaching evaluation scores.

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='eval', data=df, color='orchid', edgecolor='white')
plt.title("Age vs Teaching Evaluation")
plt.xlabel("Age")
plt.ylabel("Evaluation Score")
plt.show()
```

Q4) Using gender-differentiated scatter plots, plot the relationship between age and teaching evaluation scores.

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='eval', hue='gender', data=df, palette={'Male':'purple','Female':'pink'})
plt.title("Age vs Teaching Evaluation by Gender")
plt.xlabel("Age")
plt.ylabel("Evaluation Score")
plt.legend(title='Gender')
plt.show()
```

Q5) Create a scatterplot of age and evaluation scores, differentiated by gender and tenure

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='age', y='eval', hue='gender', style='tenure', data=df,
                palette={'Male':'purple','Female':'pink'}, s=100)
plt.title("Age vs Teaching Evaluation by Gender and Tenure")
plt.xlabel("Age")
plt.ylabel("Evaluation Score")
plt.legend(title='Gender/Tenure')
plt.show()
```
